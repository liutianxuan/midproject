**Find out whether road_conditions and light_conditions are related.**

```python
road_light = data[['month','hour','road_conditions','light_conditions']]
road_light.head()
```

```python
road_light['month'] = road_light['month'].astype(int)

import calendar
road_light['month'] = road_light['month'].apply(lambda x: calendar.month_abbr[x])
road_light.head()
```

```python
road_light['hour'] = road_light['hour'].astype(str)
road_light.dtypes
```

```python
crosstab = pd.crosstab(road_light['road_conditions'],road_light['light_conditions'], margins=False)
```

```python
crosstab
```

```python
# H0: road_conditions and light_conditions are independent
# H1: road_conditions and light_conditions are not independent

import scipy.stats as st
st.chi2_contingency(crosstab)
```

```python
road_light.corr()
```

```python
# p < 0.05, reject H0, light_conditions and road_conditions are not independent.
```
